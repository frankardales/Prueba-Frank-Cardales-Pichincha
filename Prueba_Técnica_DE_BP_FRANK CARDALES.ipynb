<b>PEDIDO:  Se solicita elaborar un reporte que calcule de forma automática la siguiente información:  Equipo, Grupo al que pertenece, Total Puntos Obtenidos en todos los partidos y Sumatoria del Gol Diferencia logrado en cada partido.

--------------------------------------------
<b>Reglas para asignación de Puntos:
--------------------------------------------
  * Si gana un partido obtendrá 3 puntos.
  * Si pierde un partido obtendrá 0 puntos.
  * Si empata un partido obtendrá 1 punto.
----------------------------------------------------------------
<b>Gol Diferencia:
----------------------------------------------------------------
  * Sumatoria del total de número de goles anotados menos número de goles recibidos.

-------------------------------------------
Fuente:  Tabla "Partidos"

```python
# Ejercicio # 1
conn = sqlite3.connect('test.db')


####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''

WITH partidos_expand AS (
    SELECT
        TRIM(SUBSTR(partido,1,INSTR(partido,'vs.')-1)) AS equipo1,
        TRIM(SUBSTR(partido,INSTR(partido,'vs.')+3)) AS equipo2,
        grupo,
        CAST(SUBSTR(resultado,1,INSTR(resultado,',')-1) AS INT) AS goles1,
        CAST(SUBSTR(resultado,INSTR(resultado,',')+1) AS INT) AS goles2
    FROM Partidos
),
equipos AS (
    SELECT equipo1 AS equipo, grupo,
        CASE WHEN goles1>goles2 THEN 3 WHEN goles1=goles2 THEN 1 ELSE 0 END puntos,
        goles1-goles2 gol_dif
    FROM partidos_expand
    UNION ALL
    SELECT equipo2 AS equipo, grupo,
        CASE WHEN goles2>goles1 THEN 3 WHEN goles1=goles2 THEN 1 ELSE 0 END puntos,
        goles2-goles1 gol_dif
    FROM partidos_expand
)
SELECT equipo, grupo,
       SUM(puntos) puntos,
       SUM(gol_dif) gol_diferencia
FROM equipos
GROUP BY equipo, grupo
ORDER BY grupo, puntos DESC;
                       ''')

for row in cursor:
  print(row)
conn.close()
```

```python
from typing import cast
import sys
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import explode, col, desc, avg, min, max, rank, asc, round, pow, sqrt
from pyspark.sql import Window

spark = SparkSession.builder\
        .master("local")\
        .appName("Colab")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()

####### PROGRAME SU RESPUESTA AQUI ##############

# 5.1 Leer archivo JSON (multiline porque el json viene en formato array)
df = spark.read.option("multiline","true").json("/content/CENTROS_EDUCATIVOS_MADRID.json")

# Revisar estructura
df.printSchema()


# Seleccionar columnas necesarias y limpiar datos
df_centros = df.select(
    col("centro_titularidad"),
    col("centro_nombre"),
    col("direccion_coor_x").cast("double"),
    col("direccion_coor_y").cast("double")
).filter(
    col("direccion_coor_x").isNotNull() &
    col("direccion_coor_y").isNotNull()
)


# 5.2 Promedio de coordenadas por titularidad
df_promedio = df_centros.groupBy("centro_titularidad").agg(
    avg("direccion_coor_x").alias("avg_x"),
    avg("direccion_coor_y").alias("avg_y")
)


# unir promedios con dataset original
df_join = df_centros.join(df_promedio, "centro_titularidad")


# 5.2.2 Calcular distancia euclidiana SIN UDF (mejor práctica en Spark)
df_dist = df_join.withColumn(
    "distancia",
    sqrt(
        pow(col("direccion_coor_x") - col("avg_x"), 2) +
        pow(col("direccion_coor_y") - col("avg_y"), 2)
    )
)


# 5.2.3 encontrar centro más cercano por grupo
window = Window.partitionBy("centro_titularidad").orderBy(col("distancia").asc())

df_result = df_dist.withColumn(
    "ranking",
    rank().over(window)
).filter(col("ranking") == 1)


df_result.select(
    "centro_titularidad",
    "centro_nombre",
    "distancia"
).show()

#################################################


df_export = df_result.select(
    col("centro_titularidad"),
    col("centro_nombre"),
    col("distancia").cast("double")
)

# PARQUET
df_export.write.mode("overwrite").parquet("/tmp/resultado_parquet")

# JSON
df_export.write.mode("overwrite").json("/tmp/resultado_json")

# CSV con separador |
df_export.write \
    .option("header", True) \
    .option("delimiter", "|") \
    .mode("overwrite") \
    .csv("/tmp/resultado_csv")

df_result.cache()
df_result.count()


spark.stop()
```
